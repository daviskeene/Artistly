```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
tobi_df = pd.read_csv('discography_tobi lou.csv')
my_df_short = pd.read_csv('user_daviskeene_medium_term.csv')
```

```python
# clean up data
cols_to_drop = ['Unnamed: 0', 'time_signature', 'release_date', 'popularity', 'tempo', 'length', '']
cols_to_augment = ['loudness']

def sigmoid(x):
    return 1/(1 + np.exp(-x))

tobi_df_new = tobi_df.drop(columns=cols_to_drop)
tobi_df_new[cols_to_augment] = tobi_df_new[cols_to_augment].apply(sigmoid)
tobi_df_new._get_numeric_data()

my_df_short_new = my_df_short.drop(columns=cols_to_drop)
my_df_short_new[cols_to_augment] = my_df_short_new[cols_to_augment].apply(sigmoid)
my_df_short_new._get_numeric_data()

# get an average of all the columns for my songs
my_mean = my_df_short_new.mean()
tobi_df_new['daviskeene_similarity'] = np.dot(tobi_df_new._get_numeric_data(), my_mean)
```

```python
tobi_df_new.sort_values(by='daviskeene_similarity', ascending=False)
```
