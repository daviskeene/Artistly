```python
def distance_vector(x, y):
    """
    Returns the cartesian distance between two vectors of size n
    """
    import math
    n = len(x)
    s = 0
    for i in range(n):
        s += x[i] ** 2 + y[i] ** 2
    return math.sqrt(s)

```
